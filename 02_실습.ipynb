```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
flight_filepath = "./dataset/flight_delays.csv"
```

```python
flight_data = pd.read_csv(flight_filepath, index_col='Month')
```

```python
flight_data
```

```python
# 에러, x값이 인덱스로 설정되어있기때문에 불러오지못함.
# sns.barplot(x='Month', y='AA', data=flight_data)

# x값을 인덱스를 불러와 설정.
sns.barplot(x=flight_data.index, y='AA', data=flight_data)
```

```python
plt.figure(figsize=(10, 6))
```

```python
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month") 
```

```python
# 막대그래프 그리기
sns.barplot(x=flight_data.index, y=flight_data['NK']) 

# 세로축 레이블 생성
plt.ylabel("Arrival delay (in minutes)")
```

### 예제 3

```python
import pandas as pd 
pd.plotting.register_matplotlib_converters() 
import matplotlib.pyplot as plt 
%matplotlib inline 
import seaborn as sns 
print("Setup Complete")
```

```python
# 읽어드릴 파일 경로 지정
cancer_b_filepath = "./dataset/cancer_b.csv"
cancer_m_filepath = "./dataset/cancer_m.csv"

# 변수에 파일에 있는 데이터 셋 로드
cancer_b_data = pd.read_csv(cancer_b_filepath, index_col="Id")
cancer_m_data = pd.read_csv(cancer_m_filepath, index_col="Id")

cancer_b_data.head()
```

```python
cancer_m_data.head()
```

```python
sns.distplot(a=cancer_b_data['Area (mean)'], label="Benign", kde=False)
sns.distplot(a=cancer_m_data['Area (mean)'], label="Malignant", kde=False)
plt.legend()
```

```python
sns.distplot(a=cancer_b_data['Area (mean)'], kde=False) #, label="Benign", kde=False
sns.distplot(a=cancer_m_data['Area (mean)'], kde=False) #, label="Malignant", kde=False)
# plt.legend()
```

```python
sns.distplot(a=cancer_b_data['Radius (mean)'], label="Benign", kde=False)
sns.distplot(a=cancer_m_data['Radius (mean)'], label="Malignant", kde=False)
plt.legend()
```

### 예제4

```python
import pandas as pd 
pd.plotting.register_matplotlib_converters() 
import matplotlib.pyplot as plt 
%matplotlib inline 
import seaborn as sns 
print("Setup Complete")
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
plt.figure(figsize=(12, 3))
sns.barplot(data=flights, x="year", y="passengers")
```

```python

plt.figure(figsize=(12, 3))
sns.lineplot(data=flights, x="year", y="passengers")
```

```python
plt.figure(figsize=(12, 3))
sns.pointplot(data=flights, x="year", y="passengers")
```

```python
plt.figure(figsize=(12, 7))
sns.scatterplot(data=flights, x="year", y="passengers", hue="month", size="passengers", sizes=(0, 300))
```

```python

plt.figure(figsize=(12, 3))
sns.boxplot(data=flights, x="year", y="passengers")
```

```python
plt.figure(figsize=(12, 3))
sns.boxplot(data=flights, x="month", y="passengers")
```

```python
plt.figure(figsize=(12, 3))
sns.violinplot(data=flights, x="year", y="passengers")
```

```python
plt.figure(figsize=(12, 3))
sns.violinplot(data=flights, x="month", y="passengers")
```

```python
plt.figure(figsize=(12, 3))
sns.swarmplot(data=flights, x="year", y="passengers")
```

```python
plt.figure(figsize=(12, 3))
sns.swarmplot(data=flights, x="month", y="passengers")
```

```python
flights_pivot = flights.pivot('month', 'year', 'passengers')
flights_pivot
```

```python
sns.heatmap(flights_pivot, annot=True, fmt="d", linewidths=.5, cmap='YlGnBu')
```

```python
plt.subplots(figsize=(12, 6))
sns.heatmap(flights_pivot, annot=True, fmt="d", linewidths=.5, cmap='YlGnBu')
```

```python
sns.clustermap(flights_pivot, annot=True, fmt="d",linewidths=.5, cmap='YlGnBu')
```

```python
sns.distplot(flights["passengers"])
```

```python
sns.pairplot(data=flights, hue='month')
```
